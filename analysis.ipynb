```python
df['conv_id'].nunique()
```

```python
grouped_by_conv_id = df.groupby('conv_id')
turns_per_conv = grouped_by_conv_id.size()
print("Sizes of each conversation: \n", turns_per_conv)

average_turns_per_conv = turns_per_conv.mean()
print("\nAverage number of turns per conversation: \n", average_turns_per_conv)

std_turns_per_conv = turns_per_conv.std()
print("Standard deviation of number of turns per conversation:", std_turns_per_conv)
```

```python
import numpy as np
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize 

sentence_counts = []
word_counts = []

for utt in df['utterance']:
    utt_sents = sent_tokenize(utt)
    sentence_counts.append(len(utt_sents))  
    words_in_utt_sentence = 0
    for utt_s in utt_sents:
        words = word_tokenize(utt_s)
        words_in_utt_sentence += len(words)
    word_counts.append(words_in_utt_sentence)

        
mean_sentences = np.mean(sentence_counts)
std_sentences = np.std(sentence_counts)
mean_words = np.mean(word_counts)
std_words = np.std(word_counts)


print("Sum of sentences are:", sum(sentence_counts))
print("Sum of words are:", sum(word_counts))
print("Mean sentences per utterance:", mean_sentences)
print("Standard deviation of sentences per utterance:", std_sentences)
print("Mean words per utterance:", mean_words)
print("Standard deviation of words per utterance:", std_words)
```

```python
# Apply lowercasing and lemmatization to count unique words, 
# in order to answer the last bullet point of Subtask 2 of 3,
# that is: •	Vocabulary size (separate for user/system).

from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()

vocab_set = set()

for utterance in df['utterance']:
    wds = tokenizer.tokenize(utterance.lower())
    for wd in wds:
        if wd.isalpha(): 
            lemma = lemmatizer.lemmatize(wd)
            vocab_set.add(lemma)

vocab_size = len(vocab_set)

print("Unique Vocabulary Size:", vocab_size)
```

### Training dataset - Impressions

```python
random_utterances = df['utterance'].sample(n=178)
for ru in random_utterances:
    print(ru)
```

```python
# Group by context (i.e. emotion label) and then count unique conversation IDs (conv_id) in each group.
conversations_per_emotion = df.groupby('context')['conv_id'].nunique()
# Count of unique conversations that have at least one utterance with that emotion.
conv_per_emo = conversations_per_emotion
print(conv_per_emo) 
print(type(conv_per_emo))
```

```python
mean_conv_per_emo = np.mean(conv_per_emo)
std_conv_per_emo = np.std(conv_per_emo)
print("Mean for conversations per emotion: ", mean_conv_per_emo)
print("Standard deviation for conversations per emotion: ", std_conv_per_emo)
```

```python
tenth_percentile = conv_per_emo.quantile(0.10)

under_represented_emotions = conv_per_emo[conv_per_emo < tenth_percentile]

print("10th Percentile value:", tenth_percentile)
print("Under-represented emotions:\n", under_represented_emotions)
```
